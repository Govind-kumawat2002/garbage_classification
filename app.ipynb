```python
dict_path={"battery":[],"biological":[],"brown-glass":[],"green-glass":[],"clothes":[],"metal":[],"paper":[],"plastic":[],"shoes":[],"trash":[],"white-glass":[],"cardboard":[]}
# Image_path=r"dict_path={"battery":[],"biological":[],"brown-glass":[],"green-glass":[],"clothes":[],"metal":[],"paper":[],"plastic":[],"shoes":[],"trash":[],"white-glass":[]}
Image_path=r"F:\Projects\garbage_classification\Data-set"
for dir in os.listdir(Image_path):
    dirt_path = os.path.join(Image_path,dir)
    print(dirt_path)
    for image_path in os.listdir(dirt_path):
        dict_path[dir].append(os.path.join(dirt_path,image_path))
```

```python
img_arr_lst=[]
lables=[]

for key,value in dict_path.items():
    print(value)
    for path in value:
        img_arr=cv2.imread(path)# it read the covert data in array
        resize_img=cv2.resize(img_arr,(150,150))
        img_arr_lst.append(resize_img)
        lables.append(key)
```

```python
array_image = np.array(img_arr_lst)
labels = np.array(lables)
array_image.shape          ,    labels.shape 
```

```python
dt={'battery':0,'biological':1,'brown-glass':2,'cardboard':3,'clothes':4,'green-glass':5,'metal':6,'paper':7,'plastic':8,'shoes':9,'trash':10,'white-glass':11}
y = []
for label in labels:
    y.append(dt[label])
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(array_image,labels,test_size=0.2,random_state=42)
```

```python
x_train = x_train/255
x_test = x_test/255
x_train.shape , y_train.shape
```

```python
# Define the CNN architecture
model = Sequential([
    # Convolutional layers
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3,3), activation='relu'),

    #flattening layer
    layers.Flatten(),

    # Dense layers for classification
    layers.Dense(64, activation='relu'),
    layers.Dense(12, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Print model summary for a detailed overview
model.summary()
```
